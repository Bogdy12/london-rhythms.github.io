```python
# Restrict table to just London
locations = locations[(locations['city'] == ' London')]
locations.head()
```

```python
# loop through table and store results from API in lon lat arrays

lon = []
lat = []

for p in locations['postcode']:
    
    loc = p.replace(" ", "")
    
    response = g.geocode(query=loc, language='en', exactly_one=True, timeout=5)
    
    lon.append(response.longitude)
    lat.append(response.latitude)
    
locations['longitude'] = lon
locations['latitude'] = lat

locations.to_csv('locations_lonlat.csv')
```

```python
# test - plot on map
londonlocs = folium.Map( location=[51.5072, -0.0275], zoom_start=11, tiles='OpenStreetMap' ) 

for row in locations.iterrows():
    
    folium.CircleMarker( location = [row[1].latitude, row[1].longitude], fill=True, fill_opacity=1, radius=2).add_to(londonlocs)
    
londonlocs
```
